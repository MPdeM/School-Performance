## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Calculate the total number of schools
total_schools = len(school_data_complete['School ID'].unique())
#Calculate the total number of students
total_students = len(school_data_complete['Student ID'].unique())
#Calculate the total budget
total_budget = school_data_complete['budget'].unique().sum()
#Calculate the average math  and reading scores
average_math = round(school_data_complete['math_score'].mean(),2)
average_reading = round(school_data_complete['reading_score'].mean(),2)
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_average = round((average_math + average_reading)/2,2)

#Calculate the percentage of students with a passing math and reading scores (70 or greater)
students_pass_math = len(school_data_complete.loc[school_data_complete['math_score'] >= 70])
students_pass_read = len(school_data_complete.loc[school_data_complete['reading_score'] >= 70])
perc_students_pass_math = round( (students_pass_math)/total_students*100,2)
perc_students_pass_read = round( (students_pass_read)/total_students*100,2)
```

```python
#Create a dataframe to hold the above results
District_Summary = pd.DataFrame({"Total Schools":[total_schools],
                                "Total Students":[total_students],
                                "Total Budget":[total_budget],
                                "Average Math Score": [average_math],
                                "Average Reading Score": [average_reading],
                                "% Passing Math":[perc_students_pass_math],
                                "% Passing Reading":[perc_students_pass_read],
                                "% Overall Passing Rate":[overall_average]})

District_Summary["% Passing Math"] = District_Summary["% Passing Math"].map("{:,.2f}%".format)
District_Summary["% Passing Reading"] = District_Summary["% Passing Reading"].map("{:,.2f}%".format)
District_Summary["% Overall Passing Rate"] = District_Summary["% Overall Passing Rate"].map("{:,.2f}%".format)
District_Summary["Total Budget"] = District_Summary["Total Budget"].map("${:,.2f}".format)
District_Summary["Total Students"] = District_Summary["Total Students"].map("{:,}".format)

District_Summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Create a subset with students that pass math and other that pass reading 
school1 = school_data_complete.loc[:,['school_name','size','budget','reading_score','math_score']]

#Calculate the percentage of students with a passing math and reading scores (70 or greater)
schoolmath = school_data_complete.loc[:,['school_name','math_score']]
schoolmath_gr= schoolmath.loc[(schoolmath['math_score'] >= 70),:].groupby(["school_name"]).count()
schoolmath_gr=schoolmath_gr.rename(columns={"math_score":"# passing math"})

schoolread = school_data_complete.loc[:,['school_name',"type",'reading_score']]
schoolread_gr= schoolread.loc[(schoolread['reading_score'] >= 70),:].groupby(["school_name"]).count()
schoolread_gr=schoolread_gr.rename(columns={"reading_score":"# passing reading"})

school_groups1 = school1.groupby(["school_name"]).mean()

school_sumary_0 = pd.merge(school_data, school_groups1, how="left", on=["school_name"])

school_sumary_1 = pd.merge(school_sumary_0, schoolmath_gr, how="left", on=["school_name"])
school_sumary_2 = pd.merge(school_sumary_1,schoolread_gr,how="left", on=["school_name"]) 

school_sumary_2["Per student Budget"] = school_sumary_2["budget_x"]/school_sumary_2["size_x"]
school_sumary_2["% Passing Math"] = round(school_sumary_2["# passing math"]/school_sumary_2["size_x"]*100,2)
school_sumary_2["% Passing Reading"] = round(school_sumary_2["# passing reading"]/school_sumary_2["size_x"]*100,2)
school_sumary_2["% Overall Passing Rate"] = round((school_sumary_2["% Passing Math"]+school_sumary_2["% Passing Reading"])/2,2)

del school_sumary_2['size_y']
del school_sumary_2['budget_y']
del school_sumary_2['School ID']
del school_sumary_2['type_y']

# create a DataFrame

school_sumary_3 = pd.DataFrame({'School Name':school_sumary_2["school_name"],"Type":school_sumary_2['type_x'],
                               "Total Students":school_sumary_2["size_x"],"Total Budget":school_sumary_2["budget_x"],
                               "Per student Budget": school_sumary_2["Per student Budget"],
                               "Average Reading Score":school_sumary_2['reading_score'],
                               "Average Math Score":school_sumary_2['math_score'],
                               '% Passing Math':school_sumary_2['% Passing Math'],
                               '% Passing Reading':school_sumary_2['% Passing Reading'],
                               '% Overall Passing Rate':school_sumary_2['% Overall Passing Rate']})  

#Formating

school_sumary_3["Total Students"]=school_sumary_3["Total Students"].map("{:,.0f}".format)
school_sumary_3["Total Budget"]=school_sumary_3["Total Budget"].map("${:,.2f}".format)
school_sumary_3["Per student Budget"]=school_sumary_3["Per student Budget"].map("${:,.2f}".format)
school_sumary_3["Average Reading Score"]=school_sumary_3["Average Reading Score"].map("{:,.2f}".format)
school_sumary_3["Average Math Score"]=school_sumary_3["Average Math Score"].map("{:,.2f}".format)
school_sumary_3["% Passing Math"]=school_sumary_3["% Passing Math"].map("{:,.2f}%".format)
school_sumary_3["% Passing Reading"]=school_sumary_3["% Passing Reading"].map("{:,.2f}%".format)
school_sumary_3["% Overall Passing Rate"]=school_sumary_3["% Overall Passing Rate"].map("{:,.2f}%".format)
School_sumary_index = school_sumary_3.set_index("School Name")
#school_sumary_3.columns
School_sumary_index
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Sort and display the top five schools in overall passing rate

upper_performers = school_sumary_3.sort_values(['% Overall Passing Rate'], ascending=False)
upper_performers = upper_performers.iloc[0:5,:]
upper_performers = upper_performers.set_index("School Name")
upper_performers
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
#Sort and display the five worst-performing schools

lower_performers = school_sumary_3.sort_values(['% Overall Passing Rate'], ascending=True)
lower_performers = lower_performers.iloc[0:5,:]

lower_performers = lower_performers.set_index("School Name")
lower_performers
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_sumary_2['Spending Ranges (Per Student)']= pd.cut(school_sumary_2['Per student Budget'], spending_bins ,labels=group_names)
school_spending_scores = school_sumary_2.groupby("Spending Ranges (Per Student)").mean()
del school_spending_scores['size_x']
del school_spending_scores['budget_x']
del school_spending_scores['Per student Budget']
del school_spending_scores['# passing math']
del school_spending_scores['# passing reading']

school_spending_scores = school_spending_scores.rename(columns={'reading_score':"Average Reading Score",
                                                                'math_score':"Average Math Score"})
#Formating

school_spending_scores["Average Reading Score"]=school_spending_scores["Average Reading Score"].map("{:,.2f}".format)
school_spending_scores["Average Math Score"]=school_spending_scores["Average Math Score"].map("{:,.2f}".format)
school_spending_scores["% Passing Math"]=school_spending_scores["% Passing Math"].map("{:,.2f}%".format)
school_spending_scores["% Passing Reading"]=school_spending_scores["% Passing Reading"].map("{:,.2f}%".format)
school_spending_scores["% Overall Passing Rate"]=school_spending_scores["% Overall Passing Rate"].map("{:,.2f}%".format)

school_spending_scores
```
